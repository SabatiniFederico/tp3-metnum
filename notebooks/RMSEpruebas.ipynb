```python
df = pd.read_csv('../data/train.csv')
df = df[['precio','habitaciones','metroscubiertos', 'banos','garages', 'antiguedad']]
# df = df[df['provincia']=='Distrito Federal'].dropna(subset = ['habitaciones','metroscubiertos', 'banos','garages', 'antiguedad'])

X = df['metroscubiertos'].values
y = df['precio'].values

X = X.reshape(len(X),1)
y = y.reshape(len(y),1)
# df.info()
df.corr()
```

```python
def errorRMSE(y1,y2):
    return sqrt ( mean_squared_error(y1,y2))

def errorRMSLE(y1,y2):
    return sqrt ( mean_squared_error(np.log(y1),np.log(y2)))

linear_regressor = metnum.LinearRegression()

linear_regressor.fit(X,y)
df['prediction'] = linear_regressor.predict(X)

errorRMSE = errorRMSE(df['precio'].values,df['prediction'].values)
errorRMSLE = errorRMSLE(df['precio'].values,df['prediction'].values)

print (df['precio'].values)
print('ERROR_RMSE: ' + str(errorRMSE))
print('ERROR_RMSLE: ' + str(errorRMSLE))
```
